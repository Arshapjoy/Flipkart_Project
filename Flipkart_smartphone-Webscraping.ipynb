```python
product_name = []
product_price = []
review_rating = []
review_count = []
relative_url = []
product_details = []


for i in range (1,51):
    
    # website in variable
    website = 'https://www.flipkart.com/search?q=smartphone&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=off&as=off' + str(i)
    
    # request
    response = requests.get(website)
    
    # soup object
    soup = BeautifulSoup(response.content, 'html.parser')
    
    # results
    results = soup.find_all('div', {'class':'_2kHMtA'})
    
    # loop through results
    for result in results:
        # name
        try:
            product_name.append(result.find('div', {'class':'_4rR01T'}).get_text()) 
        except:
            product_name.append('n/a')
    
        # price
        try:
            product_price.append(result.find('div', {'class':'_30jeq3 _1_WHN1'}).get_text().strip())
        except:
            product_price.append('n/a')
    
        # review rating
        try:
            review_rating.append(result.find('div', {'class':'_3LWZlK'}).get_text().strip())
        except:
            review_rating.append('n/a')
    
        #product details
        try:
            product_details.append(result.find('ul', {'class':'_1xgFaf'}).get_text().strip().replace('|',', '))
        except:
            product_details.append('n/a')
```

```python
product_overview = pd.DataFrame({'Name': product_name, 'Ratings-Stars':review_rating, 'Price':product_price, 'Details': product_details})
```
